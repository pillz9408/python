### .drop_duplicates() : 중복 제거
 - keep = 'first' : 기본값, 중복된 값중 제일 첫번째 값만 출력
 - keep = 'last' : 중복된 값 중 가장 마지막 값만 출력
 - keep = False : 중복된 값들을 전부 제거

```python
i = df.loc[(df.item_name=='Steak Salad')|(df.item_name=='Bowl')]
i = i.drop_duplicates('item_name')
i
```

```python
i = df.loc[(df.item_name=='Steak Salad')|(df.item_name=='Bowl')]
i = i.drop_duplicates('item_name',keep='last')
i
```

```python
f = df.loc[(df.item_name=='Steak Salad')|(df.item_name=='Bowl')]
f = f.drop_duplicates('item_name',keep=False)
f
```

```python
Dataurl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv'
Df=pd.read_csv(Dataurl, encoding='utf-8', index_col= 'Location')
Df
```

```python
df1 = Df[:4]
df2 = Df[4:]
display(df1)
display(df2)
```

```python
# df1 과 df2 를 하나의 데이터프레임으로 만들기
a = pd.concat([df1,df2])
a
```

```python
df3= Df.iloc[:2,:4]
df4= Df.iloc[:,3:]
display(df3)
display(df4)
```

```python
# df3과 df4 를 합치기
# 단, 둘다 포함되는 연도에 대해서만 고려하기(공통칼럼)
b= pd.concat([df3,df4],join='inner')
b
```

```python
# df3 과 df4 를 합치기.
# 단, 모든 칼럼 포함, 결측치는 0으로 대체
c = pd.concat([df3,df4]).fillna(0)
c
```
