```python
import matplotlib.pyplot as plt
```

```python
x = [2,3,4,5,6,7,8]
y1 = [5,7,3,9,1,8,6]
y2 = [9,5,7,2,3,7,1]

plt.plot(x, y1,label='Y1', c='r')
plt.plot(x, y2,label='Y2', c='b')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family = font_name)
```

```python
data = pd.read_csv('./Traffic_Accident_2017.csv',encoding='euc-kr')
```

```python
dfNight= data[data.주야 == '야간']
dfDay = data[data.주야 == '주간']
dfAll = data.reset_index()[['위도','경도','주야']]
```

```python
plt.scatter(x = dfNight.경도, y= dfNight.위도, alpha=0.1, c='b')
plt.scatter(x = dfDay.경도, y= dfDay.위도, alpha=0.1, c= 'r')
plt.xlabel('LON')
plt.ylabel('LAT')
plt.xlim(123,133)
plt.legend(["Night","Day"])
plt.show()
```

```python
Xn = data[data['주야']=='야간']['경도']
Yn = data[data['주야']=='야간']['위도']
Xd = data[data['주야']=='주간']['경도']
Yd = data[data['주야']=='주간']['위도']

plt.scatter(Xn,Yn,c='b',alpha=0.1, label = 'Night')
plt.scatter(Xd,Yd,c='r',alpha=0.1, label = 'Day')
plt.xlabel('LON')
plt.ylabel('LAT')
plt.legend()
plt.xlim(123,133)
plt.show()
```
