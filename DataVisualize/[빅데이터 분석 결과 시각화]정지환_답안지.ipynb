```python
import pandas as pd
import numpy as np
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family = font_name)
```

```python
data = pd.read_csv('./Traffic_Accident_2017.csv',encoding='euc-kr')
```

```python
lat = data['위도']
lon = data['경도']
```

```python
def func1(row):
    if '주간'in row:
        return 'day'
    else:
        return 'night'

```

```python
data['주야']=data['주야'].apply(func1)

```

```python

plt.scatter(x=lon,y=lat,alpha=0.1,label='day' ,c='b')
plt.scatter(x=lon,y=lat,alpha=0.1,label='night', c='r')
plt.xlabel('LON')
plt.ylabel('LAT')
plt.legend()
plt.xlim(123,133)
plt.show()
```
