# G마켓 베스트셀러

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from tqdm.notebook import tqdm
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://www.gmarket.co.kr/')
driver.maximize_window()
```

```python
from selenium.webdriver.common.action_chains import ActionChains
actions = ActionChains(driver)
```

```python
best = driver.find_element_by_xpath('//*[@id="desktop_layout-header"]/div/div/div[2]/div[2]/ul/li[1]/a')
actions.move_to_element(best)
actions.click(best)
actions.perform()
```

```python
cat_list = []
title_list = []
price_list = []

for i in tqdm(range(200)):
    img = driver.find_elements_by_css_selector('img.lazy')
    img[i].click()
    
    #상세페이지
    driver.implicitly_wait(5)  #time.sleep 과 같은 기능 웹사이트 로딩까지
    cat = driver.find_element_by_css_selector('li.on > a')
    cat_list.append(cat.text)
    price = driver.find_element_by_css_selector('strong.price_real')
    price_list.append(price.text)
    title = driver.find_element_by_css_selector('h1.itemtit')
    title_list.append(title.text)
    
    driver.back()
    time.sleep(1)
```

```python
data = {'제품명':title_list, '가격':price_list, '카테고리':cat_list}
data
```

```python
gdata = pd.DataFrame(data)
gdata.to_excel('G마켓가격.xlsx', encoding = 'utf-8-sig')
```
