# 네이버 증시에서 KOSPI  종목, 현재가격 가져오기 

```python
nc = req.get("https://finance.naver.com/sise/sise_quant.naver")
nc
```

```python

soup1 = bs(nc.text,'html.parser')
soup1
```

```python
title = soup1.select('a.tltle')
num = soup1.select('td.number')
```

```python
len(num)
len(title)
```

```python
print(title[1].text)
print(num[1].text)
```

```python
title_list=[]
for i in title:
    
    title_list.append(i.text)
```

```python
for n in num:
    print(n.text)
```

```python
num_list=[]
for n in num:
    num_list.append(n.text)
```

```python
num_list
```

```python
price_list=[]
for i in range(0,1000):
    if i % 10 == 0:
        price = num_list[i]
        price_list.append(price)
```

```python
import pandas as pd ### 판다라이브러리 = 내가 만든 리스트를 엑셀파일로 만들어주는 친구
```

```python
len(price_list)
```

```python
len(title_list)
```

```python
data = {'종목명' : title_list, '현재가': price_list}
```

```python
kospi = pd.DataFrame(data)
kospi
```

```python
kospi.to_excel('코스피지수.xlsx', encoding = 'utf-8-sig')
```

```python
from tqdm import tqdm_notebook    ### for 문 확인하는 라이브러리
```

```python
price_list=[]
for i in tqdm_notebook(range(0,1000)):
    if i % 10 == 0:
        price = num_list[i]
        price_list.append(price)
```

# 벅스 사이트에서 순위, 곡, 가수 이름 가져오기

```python
bugs = req.get("https://music.bugs.co.kr/chart")
bugs
```

```python
soup = bs(bugs.text,'html.parser')
soup
```

```python
rank = soup.select('div.ranking>strong')
title = soup.select('p.title>a')
artist = soup.select('p.artist>a')
```

```python
artist_con = soup.select('a.artistTitle')
```

```python
for i in artist_con:
    artist.remove(i)
```

```python
rankList=[]
for i in rank:
    rankList.append(i.text)
```

```python
title_lists = []
for i in title:
    title_lists.append(i.text)
```

```python
artist_list = []
for i in artist:
    artist_list.append(i.text)
```

```python
bugsDT = {'순위' : rankList, '곡명' : title_lists, '가수명' : artist_list}
bugsDT
```

```python
bugs = pd.DataFrame(bugsDT)
bugs.set_index('순위')
```

```python
bugs.to_excel('벅스랭킹.xlsx', encoding = 'utf-8-sig')
```
