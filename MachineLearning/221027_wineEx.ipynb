```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
wine_df = pd.read_csv('./wine_data.csv')
wine_df.head()
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1])
plt.title('wine data')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

### K-Means

```python
km = KMeans(3)
km_cluster = km.fit_predict(wine_df)
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1], c=km_cluster)
plt.title('wine - KMeans')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

### 계층적 군집 분석 AGG

```python
a =AgglomerativeClustering(3, linkage='ward')
a_cluster = a.fit_predict(wine_df)
```

```python
plt.scatter(wine_df.values[:,0],wine_df.values[:,1], c=a_cluster)
plt.title('wine - aggCluster')
plt.xlabel('OD_read')
plt.ylabel('Proline')
plt.show()
```

### 실루엣 지수
 - 얼마나 클러스터에 데이터가 잘 맞는지 분석해주는 지표
 - 이 값은 -1 ~ 1 까지 숫자로 나타남
 - 값이 1에 가까울수록 클러스터가 완벽하게 구분 됐다는 뜻

```python
print('실루엣 지수 : \n')
print('Kmeans : ', silhouette_score(wine_df, km_cluster))
print('Agg : ', silhouette_score(wine_df, a_cluster))
```
