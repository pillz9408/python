# 인공 신경망 

#### - 은닉층의 크기를 정의 : hidden_layer_sizes
 - 은닉층의 크기를 정의하는 hidden_layer_sizes는 은닉층의 수와 뉴런의 갯수를 튜플로 전달함
 - ex) 10개의 뉴런을 가진 2개의 은닉층을 만든다,  hidden_layer_sizes=(10,10)
 -  hidden_layer_sizes의 기본값은 (100,)
 - 하나의 은닉층에 10개의 뉴런이 있다 (10,)

#### - 활성화 함수 activation
 - 'activation' : sigmoid -> logistic
 - relu 함수 기본값

#### - 규제를 적용 : alpha
 - 일반적으로 L1규제, 효과 크지 않기 때문에, L2규제만 지원함(sklearn 한정)
 - alpha 기본값 : 0.0001 
 

#### - batch_size : 배치의 크기 (데이터를 얼마만큼 쪼개서 볼껀지)
#### - learning_rate_init: 학습률 초기값으로 지정
#### - learning_rate : 학습률 크기
#### - max_iter: 경사하강법 사용을 위한 횟수 지정


```python
import pandas as pd
import numpy as np
import mglearn
import graphviz
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

- 인공신경망에 two_moons 데이터셋을 적용
- 은닉 유닛의 갯수는 따로 설정X, 기본값 100 사용

```python
x, y = make_moons(n_samples=100, noise=0.25, random_state=18)

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=18)
```

```python
# 모델 선언, 학습
mlp = MLPClassifier(random_state=0)
mlp.fit(x_train, y_train)
# 경계선 시각화
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

- 은닉층을 1층, 은닉유닛을 10개로 적용

```python
# 모델 선언, 학습
mlp = MLPClassifier(random_state=0, hidden_layer_sizes= [10])
mlp.fit(x_train, y_train)
# 경계선 시각화
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

- 은닉 유닛 추가
- 은닉층 추가
- 활성화 함수 사용

```python
# 10개의 유닛으로 구성된 두 개의 은닉층
# 모델 선언, 학습
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(x_train, y_train)
# 경계선 시각화
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```

```python
# 활성화 함수 적용
# 모델 선언, 학습
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=[10,10], activation='tanh')
mlp.fit(x_train, y_train)
# 경계선 시각화
mglearn.plots.plot_2d_separator(mlp, x_train, fill=True, alpha=.3)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.xlabel('property 0')
plt.ylabel('property 1')
plt.show()
```
