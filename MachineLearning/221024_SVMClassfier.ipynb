```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
```

### 1.문제정의(목적) - SVM으로 iris 데이터를 분류 해보자

### 2. 데이터 수집

```python
#아이리스 데이터 불러오기
df = load_iris()
df.keys()
```

```python
#독립변수, 종속변수 설정
x = df.data
y = df.target
```

```python
#데이터 프레임화
x = pd.DataFrame(x, columns=df.feature_names)
y = pd.DataFrame(y, columns=['species'])
```

### 3. 데이터 전처리
 - 스케일링
 - 레이블 인코딩

```python
#스케일링
transformer = MinMaxScaler()
transformer.fit(x)
```

```python
x_data = transformer.transform(x)
```

```python
x_data
```

```python
# 레이블 인코딩
le = LabelEncoder()
le.fit(y)
```

```python
print(le.classes_)
```

```python
y_data = le.transform(y)
```

```python
y_data
```

```python
# train, test split
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data, random_state = 0)
```

### 4. EDA(탐색적 데이터 분석) - 클래스별 비율, 시각화

### 5. 모델 선택

```python
model = SVC(kernel='rbf', C=1.0, gamma='auto', probability=True) #가우시안 커널
```

### 6. 모델 학습

```python
model.fit(x_train, y_train)
```

### 7. 모델 평가

```python
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

```python
labels = ['setosa', 'versicolor', 'verginica']
```
