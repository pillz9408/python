### 이용이력 데이터 분석

```python
ul.head()
```

회원이 스포츠 센터를 이용할 때 시간이 찍히기 때문에 시간적 요소를 분석해 볼 수있다.

<예시>
- 한달 이용 횟수의 변화
- 회원이 스포츠센터를 정기적으로 이용하는지
- 회원들이 월별 스포츠센터를 이용하는 평균수치, 중앙값, 최대/최소값

```python
ul['usedate'] = pd.to_datetime(ul['usedate']) #데이터를 날짜형으로 변환
```

```python
ul['YM'] = ul['usedate'].dt.strftime('%Y%m')
```

```python
ul
```

```python
# 월별, 고객 id별 이용 횟수 카운트
ul_month = ul.groupby(['YM','customer_id'], as_index=False).count()
ul_month
```

#### 월별 고객의 사용횟수 데이터

```python
#컬럼 이름 변경
ul_month.rename(columns={'log_id':'cnt'}, inplace=True)
#usedate컬럼 삭제
ul_month.drop('usedate', axis=1, inplace=True)
ul_month.tail()
```

```python
#고객별 월평균 사용횟수
ul_cust = ul_month.groupby('customer_id')['cnt'].agg([np.mean, np.median, np.max, np.min])
ul_cust = ul_cust.reset_index(drop=False)
ul_cust.head()
```

#### 정기적으로 스포츠센터를 사용하는 고객인지 확인

- 매주 같은 요일에 스포츠센터에 왔는지 아닌지로 판단
- 월/요일별로 방문한 횟수의 최대값이 4이상인 요일이 하나라도 있다면, 이 회원은 정기 사용자로 판단

```python
#요일 뽑기
# weekday: 월요일 - 0, 일요일 - 6 
ul['weekday'] = ul['usedate'].dt.weekday
ul
```

```python
ul_week = ul.groupby(['customer_id','YM','weekday'], as_index=False).count()
ul_week
```

```python
#컬럼 이름 변경
ul_week.rename(columns={'log_id':'cnt'}, inplace=True)
#usedate컬럼 삭제
del ul_week['usedate']
ul_week.tail()
```

```python
ul_week.head()
```

-> 토요일마다 정기적으로 방문하는 사람인가?

ul_week 데이터프레임을 가지고 cnt4 이상이면 1, 아니면 0인 컬럼 만들기

```python
#한달에 최대값이 4이상인 요일값이 있다면 정기 방문자로 .max()로 월마다 하나의 값만
ul_week = ul_week.groupby('customer_id', as_index=False)[['cnt']].max()
ul_week
```

```python
# 방법1 : apply (lambda)
ul_week['flag1'] = ul_week['cnt'].apply(lambda x : 1 if x>=4 else 0)
```

```python
ul_week
```

```python
# 방법 2: boolean indexing
ul_week['flag2'] = 0
is_over4 = (ul_week['cnt']>=4)
ul_week.loc[(is_over4), 'flag2']= 1
ul_week
```

```python
#방법 3: where문 사용
ul_week['flag3'] = 0
# 조건이 참이면 0, 거짓이면 1을 넣음
ul_week['flag3'] = ul_week['flag3'].where(ul_week['cnt']<4, 1)
ul_week.head() 
```

#### 회원기간 계산

```python
cust_join['cal_date'] = cust_join['end_date']
cust_join.head()
```

```python
# end_date가 비워진 경우에는  cal_date에 2019-04-30으로 임의로 채워주기
cust_join['cal_date'].fillna(pd.to_datetime('20190430'),inplace=True)
cust_join.head()
```

```python
#datetime 형식으로 변환
cust_join['start_date'] = pd.to_datetime(cust_join['start_date'])
cust_join['end_date'] = pd.to_datetime(cust_join['end_date'])
cust_join['cal_date'] = pd.to_datetime(cust_join['cal_date'])
```

```python
#회원기간을 계산한 'mem_period' 컬럼 추가
cust_join['mem_period'] = (cust_join['cal_date'] - cust_join['start_date'])/np.timedelta64(1,'M')
```
